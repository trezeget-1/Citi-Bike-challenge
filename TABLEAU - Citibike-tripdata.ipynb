# Read the CSV

```python
citibike_file = pd.read_csv("citybike_df_202001_202101.csv", index_col= 0)
```

```python
citibike_file.tail()
```

```python
citibike_file.loc[citibike_file["end station id"] == 3475]
```

```python
start_stations = citibike_file["start station id"].unique()
len(start_stations)
```

```python
ending_stations = citibike_file["end station id"].unique()
len(ending_stations)
```

```python
citibike_file.loc[citibike_file["start station id"] == start_stations[1]]["start station longitude"].unique()[0]
```

```python
start_station_latitude = []
start_station_longitude = []

end_station_latitude = []
end_station_longitude = []

for x in range(len(start_stations)):
    unique_start_station_latitude = citibike_file.loc[citibike_file["start station id"] == start_stations[x]]["start station latitude"].unique()[0]
    unique_start_station_longitude = citibike_file.loc[citibike_file["start station id"] == start_stations[x]]["start station longitude"].unique()[0]
    
    start_station_latitude.append(unique_start_station_latitude)
    start_station_longitude.append(unique_start_station_longitude)
    
for x in range(len(ending_stations)):
    unique_end_station_latitude = citibike_file.loc[citibike_file["end station id"] == ending_stations[x]]["end station latitude"].unique()[0]
    unique_end_station_longitude = citibike_file.loc[citibike_file["end station id"] == ending_stations[x]]["end station longitude"].unique()[0]
    
    end_station_latitude.append(unique_end_station_latitude)
    end_station_longitude.append(unique_end_station_longitude)
```

```python
start_stations_df = pd.DataFrame({
    "Station ID" : start_stations,
    "latitude" : start_station_latitude,
    "longitude": start_station_longitude
})

start_stations_df.sort_values(by = "Station ID")
```

```python
ending_stations_df = pd.DataFrame({
    "Station ID" : ending_stations,
    "latitude" : end_station_latitude,
    "longitude": end_station_longitude
})

ending_stations_df.sort_values(by = "Station ID")
```

```python
ending_stations_df.loc[ending_stations_df["Station ID"] == 3475]
```

```python
full_list_of_stations = start_stations_df.append(ending_stations_df)

full_list_of_stations.sort_values(by = "Station ID")
```

```python
full_list_of_stations = full_list_of_stations.drop(full_list_of_stations.loc[full_list_of_stations["Station ID"] == 3475,:].index)

full_list_of_stations = full_list_of_stations[~full_list_of_stations.duplicated(subset=['Station ID'])]

full_list_of_stations = full_list_of_stations.reset_index()

del full_list_of_stations["index"]

full_list_of_stations
```

```python
full_list_of_stations.to_csv("citybike_df_list_of_stations.csv")
```
